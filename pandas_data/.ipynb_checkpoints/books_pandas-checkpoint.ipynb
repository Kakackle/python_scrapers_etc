# Amazon CS boooks dataset data visualization etc exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv('./books2.csv', parse_dates=True)
books.head()
```

```python
len(books['edition'].unique())
```

```python
cols = ['rating', 'price']
# books.loc[:, cols].plot()
books[cols].plot.box()
```

```python
books.plot.scatter(x='rating', y='price', alpha=0.5)
```

```python
# books.plot.pie(y='edition', legend=True, x='edition')
edition_counts = books['edition'].value_counts()
edition_counts = pd.DataFrame(edition_counts)
edition_counts = edition_counts.reset_index()
edition_counts.columns = ['edition', 'count']
edition_counts.head()
```

```python
fig =  edition_counts.plot.pie(
             y='count',
             x='edition')
fig.show()
```
