```python
df = pd.read_csv("../scraping_results/combined/python/python_2023-11-28.csv", index_col = 0)
# df_clean = df.dropna()
companies = df['company']
companies = companies.dropna()
companies.head()
```

```python
colors = sns.color_palette('Spectral')
```

```python
companies_count = companies.value_counts()
fig = companies_count[:20].plot.pie(colors = colors, autopct='%.0f%%')
fig.set_ylabel('')
fig.set_title('Top 20 companies pie chart')
```

```python
fig.get_figure().savefig('pie.png')
```
